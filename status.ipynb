```python
import requests

import xml.etree.ElementTree as ET

resp = requests.get("https://conda-static.anaconda.org/conda-forge/rss.xml")

import io
import dateutil

tree = ET.parse(io.BytesIO(resp.content))
channel = tree.getroot()[0]

print(dateutil.parser.parse(channel.findall('pubDate')[0].text).astimezone(), '  ', 'rss PUB')
print()

for x in reversed(tree.getroot()[0][-100:]):
    pkg = x[0].text
    
    #if ('python' not in pkg) and ('running sync' not in pkg):
    #    continue
        
    print(dateutil.parser.parse(x[-1].text).astimezone(), '  ', pkg)
```
